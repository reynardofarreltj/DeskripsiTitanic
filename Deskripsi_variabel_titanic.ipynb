```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data extracted directly from the correlation matrix in your notebook.
correlation_data = {
    'survived':    [1.00, 0.02, -0.20],
    'family_size': [0.02, 1.00, -0.69],
    'is_alone':    [-0.20, -0.69, 1.00]
}

# Define the labels for the rows/columns of the heatmap
index_labels = ['survived', 'family_size', 'is_alone']

# Create a DataFrame from the correlation data
correlation_matrix_new = pd.DataFrame(correlation_data, index=index_labels)

# Set the figure size for better readability
plt.figure(figsize=(6, 4))

# Create the heatmap using seaborn
# annot=True displays the correlation values on the map
# cmap='coolwarm' sets the color scheme
# fmt=".2f" formats the numbers to two decimal places
sns.heatmap(correlation_matrix_new, annot=True, cmap='coolwarm', fmt=".2f")

# Add a title to the plot
plt.title("Correlation Heatmap of New Variables vs. Survival")

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Membuat variabel 'family_size'
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1

# Membuat visualisasi bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='family_size', y='survived', data=titanic, palette='viridis')
plt.title('Tingkat Keselamatan Berdasarkan Ukuran Keluarga')
plt.xlabel('Ukuran Keluarga')
plt.ylabel('Tingkat Keselamatan (Survival Rate)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the titanic dataset and create new variables
titanic = sns.load_dataset('titanic')
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1
titanic['is_alone'] = 0
titanic.loc[titanic['family_size'] == 1, 'is_alone'] = 1


# Membuat visualisasi bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='is_alone', y='survived', data=titanic, palette='cool')
plt.title('Tingkat Keselamatan: Sendiri vs Bersama Keluarga')
plt.xlabel('Status Perjalanan')
plt.ylabel('Tingkat Keselamatan (Survival Rate)')
plt.xticks(ticks=[0, 1], labels=['Bersama Keluarga', 'Sendiri'])
plt.show()
```
